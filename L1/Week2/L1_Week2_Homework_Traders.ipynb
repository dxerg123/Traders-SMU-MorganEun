### Exercise 1: Data Collection and Basic Operations
### Topics Used: For Loops, Indexing, Lists, Operations, Variables

#### 1. You are given a list of daily closing prices for a stock over 5 days: [100.5, 102.3, 99.8, 103.7, 101.2].

#### 2. Store this list in a variable called stock_prices.
#### 3. Calculate the daily returns for each day (return = (current_price - previous_price) / previous_price) and store them in a new list called daily_returns.
#### 4. Print the daily_returns list.

```python
# Submit Code Here 
stock_prices = [100.5, 102.5, 99.8, 103.7, 101.2]

daily_returns = []

for i in range(1, len(stock_prices)):
    
    return_value = (stock_prices[i] - stock_prices[i - 1]) / stock_prices[i - 1]
   
   daily_returns.append(return_value)


print(daily_returns)
```

### Exercise 3: Loops and Trading Signals

#### Using stock_prices, create signals ("Buy" if >2% increase, "Sell" if <-2% decrease, "Hold" otherwise).
#### Break on "Sell".
#### Print signals and break message.

```python
# Submit Code Here 

# Initialize the signals list
signals = []

# Iterate over the stock prices; using stock_prices, create signals ("Buy" if >2% increase, "Sell" if <-2% decrease, "Hold" otherwise)
for i in range(1, len(stock_prices)):
    
    daily_return = (stock_prices[i] - stock_prices[i - 1]) / stock_prices[i - 1]
    
    if daily_return > 0.02:
        signals.append("Buy")
   
    elif daily_return < -0.02:
        signals.append("Sell")

        break  # Break on "Sell"

    else:
        signals.append("Hold")

# Print signals and break message

break_message = "Breaking on 'Sell' signal." if "Sell" in signals else "No 'Sell' signal encountered."

print("Signals:", signals)
print(break_message)
```

### Final Challenge Exercise 5: Object-Oriented Programming - Advanced Trading Strategy Class

#### Build an advanced TradingStrategy class to simulate a stock trading system with detailed tracking and reporting. Follow these steps:

#### 1. Set up the class with:
#### &nbsp;A name for the strategy.
#### &nbsp;A list for stock prices.
#### &nbsp;A list for trading signals.
#### &nbsp;A list to record trades.
#### &nbsp;Variables for cash (start with $10,000) and shares (start with 0).
#### &nbsp;A fixed number of shares to trade each time (e.g., 10).
#### 2. Create a method to add a new stock price to the class.
#### 3. Create a method to generate trading signals:
#### &nbsp;Compare each price to the previous one.
#### &nbsp;Use a 2% increase for "Buy," a 2% decrease for "Sell," and anything else for "Hold."
#### 4. Create a method to execute trades based on signals:
#### &nbsp;For "Buy," purchase shares if cash allows, updating cash and shares.
#### &nbsp;For "Sell," sell all shares if any are owned, updating cash and shares.
#### &nbsp;Record each trade as a tuple with price, action, and shares traded.
#### 5. Create a method to calculate the total portfolio value using the latest price.
#### 6. Create a method to provide a summary including:
#### &nbsp;Strategy name, signals, trades, final cash, shares owned, and total portfolio value.
#### 7. Test your class:
#### &nbsp;Use the prices [100.5, 102.3, 99.8, 103.7, 101.2].
#### &nbsp;Add each price, generate signals, execute trades, and print the summary.


```python
# Submit Code Here 

class TradingStrategy:
    def __init__(self, sample): 
        self.sample = sample # A name for the strategy
        self.stock_prices = [] # A list for stock prices
        self.trading_signals = [] # A list for trading signals
        self.trades = [] ## A list to record trades
        self.cash = 10000  # Variables for cash (start with $10,000)
        self.shares = 0  # Shares (start with 0)
        self.shares_to_trade = 10  # A fixed number of shares to trade each time (e.g., 10)

    def add_stock_price(self, price): #### Create a method to add a new stock price to the class.
        
        self.stock_prices.append(price)

    def generate_signals(self): #### Create a method to generate trading signals
    
        if len(self.stock_prices) < 2:
            return  # Not enough data to generate signals

        prev_price = self.stock_prices[-2]
        current_price = self.stock_prices[-1]
        
        if current_price > prev_price * 1.02:
            self.trading_signals.append("Buy") # For "Buy," purchase shares if cash allows, updating cash and shares.
        elif current_price < prev_price * 0.98:
            self.trading_signals.append("Sell") # For "Sell," sell all shares if any are owned, updating cash and shares.
        else:
            self.trading_signals.append("Hold")

    def execute_trades(self):
        
        if not self.trading_signals:
            return  # No signals to act on

        latest_signal = self.trading_signals[-1]
        latest_price = self.stock_prices[-1]

        if latest_signal == "Buy" and self.cash >= latest_price * self.shares_to_trade:  # For "Buy," purchase shares if cash allows, updating cash and shares.
            self.cash -= latest_price * self.shares_to_trade
            self.shares += self.shares_to_trade
            self.trades.append((latest_price, "Buy", self.shares_to_trade))
        elif latest_signal == "Sell" and self.shares > 0: # For "Sell," sell all shares if any are owned, updating cash and shares.
            self.cash += latest_price * self.shares
            self.trades.append((latest_price, "Sell", self.shares)) 
            self.shares = 0

    def calculate_portfolio_value(self):
       
        if not self.stock_prices:
            return self.cash  # No stock prices, portfolio value is just cash

        latest_price = self.stock_prices[-1]
        return self.cash + (self.shares * latest_price) # Create a method to calculate the total portfolio value using the latest price.

    def summary(self):
       
        portfolio_value = self.calculate_portfolio_value()
        return {
            # Create a method to provide a summary including strategy name, signals, trades, final cash, shares owned, and total portfolio value
            "Strategy Name": self.sample,
            "Signals": self.trading_signals,
            "Trades": self.trades,
            "Final Cash": self.cash,
            "Shares Owned": self.shares,
            "Total Portfolio Value": portfolio_value
        }

# Test the TradingStrategy class
strategy = TradingStrategy("Advanced Trading Strategy")

# Stock prices to test
test_prices = [100.5, 102.3, 99.8, 103.7, 101.2] # Use the prices [100.5, 102.3, 99.8, 103.7, 101.2].

# Add each price, generate signals, and execute trades
for price in test_prices:
    strategy.add_stock_price(price) # Add each price,
    strategy.generate_signals() # Generate signals
    strategy.execute_trades() # Execute trades

# Print the summary
print(strategy.summary()) # Print the summary
```
